```python
import pandas as pd

#Importation du fichier CSV
df = pd.read_csv("Automobiles.csv")
#Affichage des 5 premières lignes
df.head()
```

```python
#Affichage du type de données
print (df.dtypes)
```

```python
#Affichage des valeurs (il n'y en a aucune)
df.isna().sum()
```

```python
#Affichage des détails de toutes les voitures de marque Toyota
Toyota = df[df["company"] == 'toyota']
print(Toyota)
```

```python
#Le nom de la compagnie, le nombre de chevaux et prix
expensive = df[['company','horsepower','price']][df['price']==df['price'].max()]
print (expensive)
```

```python
#Affichage du nombre de voitures par comapgnie
df['company'].value_counts()
```

```python
#la voiture avec le prix le plus élevé pour chaque compagnie
max_price = df.groupby('company')['price'].max()
print (max_price)
```

```python
#la moyenne de kilométrage par compagnie
avg_km = df.groupby('company')['average-mileage'].mean()
print (avg_km)
```

```python
#Tri des voitures par prix
df.sort_values(by='price')
```
